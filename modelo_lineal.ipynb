```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
#Dataset: tamaño en miles de pies cuadrados y precio en miles de dolares
x = np.array([1,2,3]) #tamaño casa (1.000, 2.000, 3.000 pies cuadrados)
y = np.array([300,500,700])#Precio (300k, 500k, 700k dólares)
```

```python
#Predicción de una sola variable ƒ w,b ( x(i) ) = wx(i)+ b
def predict(x,w,b):
    return w*x+b
```

```python
#Función de Costo MSE
def MSE(x,y,w,b):
    m = len(x)
    return (1/2*m)*np.sum((predict(x,w,b)-y)**2)
```

```python
#función de gradiente descendente
def gradient_descent(x,y,w,b,alpha,iterations):
    m = len(x)
    for i in range(iterations):
      #calcular gradiente
      dw = (1/m)*np.sum((predict(x,w,b)-y)*x)
      db = (1/m)*np.sum(predict(x,w,b)-y)
      #actualizar parámetros
      w = w - alpha*dw
      b = b - alpha*db
      #calcular costo
      cost = MSE(x,y,w,b)
      print(f"Iteración {i + 1}: Costo {cost:.4f}, w {w:.4f}, b {b:.4f}")
    return w,b
```

```python
#inicialización de parámetros
w = 0 #peso
b = 0# bias
alpha = 0.01 #learning rate
iterations = 1000
```

```python
#entrenamiento del modelo
w,b = gradient_descent(x,y,w,b,alpha,iterations)
```

```python
print(f"Modelo entrenado: w = {w}, b = {b}")
```

```python
#predicciones para los datos
y_pred = predict(x,w,b)
```

```python
y_pred
```

```python
# Gráfico de los valores reales vs los predichos
plt.scatter(x, y, color="blue", label="Datos reales")  # Datos reales
plt.plot(x, y_pred, color="red", label="Modelo lineal")  # Línea de regresión
plt.xlabel("Tamaño de la casa (miles de pies cuadrados)")
plt.ylabel("Precio de la casa (miles de dólares)")
plt.title("Predicción del modelo vs Valores reales")
plt.legend()
plt.show()
```

```python
#Predicción nueva de una casa de 4 mil pies cuadrados
x_nueva = 4
y_pred_nueva = predict(x_nueva,w,b)
print(f"Predicción para x = {x_nueva}: y = {y_pred_nueva:.4f}")
```
